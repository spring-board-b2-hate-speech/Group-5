```python
import os
import pandas as pd
import re
import string

# Verify the directory and list files
directory = "C:/Users/suman/Documents/"
print("Current Working Directory:", os.getcwd())
print("Files in directory:", os.listdir(directory))

# Define the clean function
def clean(text):
    text = re.sub(r'https?://\S+|www.\S+', '', text)  # Use raw string literals
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)  # Updated regex pattern to remove alphanumeric words
    text = text.split()
    text = " ".join(text)
    return text

# Adjust the file path as necessary
file_path = "C:/Users/suman/Documents/HateSpeechData.csv"  # Update with your actual file path

# Read the CSV file if it exists
if os.path.exists(file_path):
    data = pd.read_csv(file_path, on_bad_lines='skip')
    # Apply the clean function
    data["tweet"] = data["tweet"].apply(clean)
    print(data.head())
else:
    print(f"File not found: {file_path}")
```
